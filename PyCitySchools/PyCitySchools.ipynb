```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
schools_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the results for the Dataframe
total_schools = schools_data_complete["School ID"].nunique()
total_students = schools_data_complete["Student ID"].nunique()
total_budget = schools_data_complete["budget"].sum()
average_math_score = schools_data_complete["maths_score"].mean()
average_reading_score = schools_data_complete["reading_score"].mean()
passing_math_percentage = (schools_data_complete["maths_score"] >= 50).mean() * 100
passing_reading_percentage = (schools_data_complete["reading_score"] >= 50).mean() * 100
overall_passing_percentage = ((schools_data_complete["maths_score"] >= 50) & (schools_data_complete["reading_score"] >= 50)).mean() * 100

# Create a DataFrame for the results
lgasummary_data = {"Total Schools": [total_schools],
                "Total Students": [total_students],
                "Total Budget": [total_budget],
                "Average Math Score": [average_math_score],
                "Average Reading Score": [average_reading_score],
                "% Passing Math": [passing_math_percentage],
                "% Passing Reading": [passing_reading_percentage],
                "% Overall Passing": [overall_passing_percentage]}
                 
lgasummary_df = pd.DataFrame(lgasummary_data)

# Format the results
lgasummary_df["Total Students"] = lgasummary_df["Total Students"].map("{:,}".format)
lgasummary_df["Total Budget"] = lgasummary_df["Total Budget"].map("${:,.2f}".format)
lgasummary_df["Average Math Score"] = lgasummary_df["Average Math Score"].map("{:.1f}".format)
lgasummary_df["Average Reading Score"] = lgasummary_df["Average Reading Score"].map("{:.1f}".format)
lgasummary_df["% Passing Math"] = lgasummary_df["% Passing Math"].map("{:.1f}%".format)
lgasummary_df["% Passing Reading"] = lgasummary_df["% Passing Reading"].map("{:.1f}%".format)
lgasummary_df["% Overall Passing"] = lgasummary_df["% Overall Passing"].map("{:.1f}%".format)

# Display the results
lgasummary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create a list to store the data for each school
school_summaries = []

# Look through each school
for school_name in school_data["school_name"]:
    # Filter students for each school
    school_students = student_data[student_data["school_name"] == school_name]

    #Calculate the results for the Dataframe
    total_students = len(school_students)
    total_budget = school_data[school_data["school_name"] == school_name]["budget"].values[0]
    average_math_score = school_students["maths_score"].mean()
    average_reading_score = school_students["reading_score"].mean()
    passing_math_percentage = (school_students["maths_score"] >= 50).mean() * 100
    passing_reading_percentage = (school_students["reading_score"] >= 50).mean() * 100
    overall_passing_percentage = ((school_students["maths_score"] >= 50) & (school_students["reading_score"] >= 50)).mean() * 100

    # Create a dictionary for the data
    school_summary = {"School Name": school_name,
                      "Total Students": total_students,
                      "Total School Budget": total_budget,
                      "Average Math Score": average_math_score,
                      "Average Reading Score": average_reading_score,
                      "% Passing Math": passing_math_percentage,
                      "% Passing Reading": passing_reading_percentage,
                      "% Overall Passing": overall_passing_percentage}
   

    school_summaries.append(school_summary)

# Create a DataFrame for the results
school_overview_df = pd.DataFrame(school_summaries)

# Format the results
school_overview_df["Total Students"] = school_overview_df["Total Students"].map("{:,}".format)
school_overview_df["Total School Budget"] = school_overview_df["Total School Budget"].map("${:,.2f}".format)
school_overview_df["Average Math Score"] = school_overview_df["Average Math Score"].map("{:.1f}".format)
school_overview_df["Average Reading Score"] = school_overview_df["Average Reading Score"].map("{:.1f}".format)
school_overview_df["% Passing Math"] = school_overview_df["% Passing Math"].map("{:.1f}%".format)
school_overview_df["% Passing Reading"] = school_overview_df["% Passing Reading"].map("{:.1f}%".format)
school_overview_df["% Overall Passing"] = school_overview_df["% Overall Passing"].map("{:.1f}%".format)

# Display the results
school_overview_df
```

```python
# Sort by Overall Passing % in descending order and create a DataFrame for the top five schools
top_schools = school_overview_df.sort_values(by="% Overall Passing", ascending=False).head(5)

# Display the results
top_schools
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a DataFrame for each year level
math_scores = pd.DataFrame()

for year in [9, 10, 11, 12]:
    condition = student_data["year"] == year
    year_series = student_data[condition].groupby("school_name")["maths_score"].mean()
    math_scores[f"Year {year} Average Math Score"] = year_series

# Reset the index to include school name
math_scores.reset_index(inplace=True)

# Format the results
math_scores["Year 9 Average Math Score"] = math_scores["Year 9 Average Math Score"].map("{:.1f}".format)
math_scores["Year 10 Average Math Score"] = math_scores["Year 10 Average Math Score"].map("{:.1f}".format)
math_scores["Year 11 Average Math Score"] = math_scores["Year 11 Average Math Score"].map("{:.1f}".format)
math_scores["Year 12 Average Math Score"] = math_scores["Year 12 Average Math Score"].map("{:.1f}".format)

# Display the results
math_scores
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Define Spending Data
spending_data = schools_data_complete

# Calculate the spending per student
spending_data["Spending per Student"] = spending_data["budget"] / spending_data["size"]

# Define the spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a new column for spending ranges
spending_data["Spending Ranges"] = pd.cut(spending_data["Spending per Student"], spending_bins, labels=labels)

# Group the data by spending ranges
spending_groups = spending_data.groupby("Spending Ranges")

# Calculate the average reading and math score
spending_summary = spending_groups.agg({
    "maths_score": "mean",
    "reading_score": "mean",
})

# Calculate the percentage passing math and reading
spending_summary["% Passing Math"] = spending_groups["maths_score"].apply(lambda x: (x >= 50).mean() * 100)
spending_summary["% Passing Reading"] = spending_groups["reading_score"].apply(lambda x: (x >= 50).mean() * 100)

# Calculate Overall Passing Rate
spending_summary["Overall Passing Rate"] = (spending_summary["% Passing Math"] + spending_summary["% Passing Reading"]) / 2

# Rename the columns
spending_summary = spending_summary.rename(columns={
    "maths_score": "Average Math Score",
    "reading_score": "Average Reading Score",
})

# Format the results
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].map("{:.1f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:.1f}".format)
spending_summary["% Passing Math"] = spending_summary["% Passing Math"].map("{:.1f}%".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:.1f}%".format)
spending_summary["Overall Passing Rate"] = spending_summary["Overall Passing Rate"].map("{:.1f}".format)

# Display the results
spending_summary
```
